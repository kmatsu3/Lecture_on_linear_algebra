```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#範囲と間隔の設定
fig = plt.figure()
ax = Axes3D(fig)


LX, LY, LZ = 1.5,1.5,1.5  # xyzメッシュのパラメータ
gridwidth=2.0 # 

ax.quiver(0, 0, 0, 1, 0, 1, color='blue', pivot='tail', length=np.sqrt(1**2+1**2+1), arrow_length_ratio=0.2)
ax.text(1,0,1,"a",fontsize=20, color='blue')
ax.quiver(0, 0, 0, 1, -1, 0, color='green', pivot='tail', length=np.sqrt(1**2+1**2+1), arrow_length_ratio=0.2)
ax.text(1,-1,0,"b",fontsize=20, color='green')
ax.quiver(0, 0, 0, 1, 1, -1, color='red', pivot='tail', length=np.sqrt(1**2+1**2+1**2), arrow_length_ratio=0.05)
ax.text(1,1,-1,"axb",fontsize=20, color='red')

ax.set_xlim([-LX, LX])
ax.set_ylim([-LY, LY])
ax.set_zlim([-LZ, LZ])
ax.set_xlabel('x',fontsize =20)
ax.set_ylabel('y',fontsize =20)
ax.set_zlabel('z',fontsize =20)
plt.tick_params(labelsize = 20)
ax.set_aspect('equal')
ax.view_init(20.0, -35.0)
plt.savefig('E1-3Ans.eps')

plt.show()
```
